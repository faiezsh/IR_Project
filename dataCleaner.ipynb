```python
import nltk
from nltk.stem import WordNetLemmatizer
from collections import defaultdict
import pandas as pd
import trie
from textblob import TextBlob
myTrie =trie.PrefixTree()
sentence_without_prepositions=[]
ids=[]
df = pd.read_excel("antique_train_que.xlsx")
i=0
punctuation = '''!()-[]{};:'"\,<>./?@#$%^&*_~'''
for index in df: 
    df[index].dropna(inplace=True)
    df[index] = df[index].astype(str)
    sentence=df[index]
    if i==0:
        ids=sentence
        i+=1
    #print (sentence)
    else :
        for s in sentence:
            s = s.lower()
            lemmatizer = WordNetLemmatizer()
            words = nltk.word_tokenize(s)
            tags = nltk.pos_tag(words)
            notTags=["SYM","PRP$","CC","PRP","TO","IN","DT","EX","FW","MD","PDT","POS","RP","UH","WDT","WP","WP$","WRB","$","(",")",",","--",".",":"]
            adv=["JJR","JJS"]
            verb=["VB","VBD","VBG","VBN","VBP","VBZ"]
            adverb=["RB","RBR","RBS"]
            nouns=["NN","NNP","NNPS","NNS"]
            exact=False
            last=[]
            for word, tag in tags:
                if word in punctuation:
                    word= word.replace(s, " ")
                if tag=="``":
                    exact=!exact 
                    last.append(word)
                elif exact:
                    last.append(word)
                elif tag in adv:
                    last.append(lemmatizer.lemmatize(word, pos="a"))
                elif tag in verb:
                    last.append(lemmatizer.lemmatize(word, pos="v"))
                elif tag in adverb:
                    last.append(lemmatizer.lemmatize(word, pos="r"))
                elif tag in nouns:
                    last.append(lemmatizer.lemmatize(word, pos="n"))
                elif tag not in notTags:
                    last.append(lemmatizer.lemmatize(word, pos="s"))
            sentence_without_prepositions.append(" ".join(last))
data=[]
for i,s in zip(ids,sentence_without_prepositions):
    data.append((i,s))
writeData = pd.DataFrame(data, columns=["id","Text"])
writeData.to_excel("newTrainqury.xlsx",index=False)

def textblo (word):
    b = TextBlob(word)
    b=str(b.correct)
    return b

```
